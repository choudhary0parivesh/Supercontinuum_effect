```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file into a DataFrame
file_path = 'background.txt'
df = pd.read_csv(file_path, header=None, delim_whitespace=True)

# Extract the columns
x_values = df.iloc[:, 0]
y_values = df.iloc[:, 1]

# Increase the size of the graph
plt.figure(figsize=(8, 4))

# Plot the data with y-axis in log scale
plt.plot(x_values, y_values, linestyle='-', color='b')
plt.yscale('log')  # Set y-axis to log scale
plt.ylim(0, 1)
plt.title('Background noise')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (AU)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the first data file
file_path1 = 'background.txt'
df1 = pd.read_csv(file_path1, header=None, delim_whitespace=True)

# Extract the columns from the first file
x_values1 = df1.iloc[:, 0]
y_values1 = df1.iloc[:, 1]

# Read the second data file
file_path2 = '90_10_delta.txt'
df2 = pd.read_csv(file_path2, header=None, delim_whitespace=True)

# Extract the columns from the second file
x_values2 = df2.iloc[:, 0]
y_values2 = df2.iloc[:, 1]

# Subtract corresponding values from the second file
y_values_diff = -y_values1 + y_values2

# Increase the size of the graph
plt.figure(figsize=(8, 4))

# Plot the subtracted values
plt.plot(x_values1, y_values_diff, linestyle='-', color='b')
plt.yscale('log')
plt.ylim(0, 1)
plt.title("Spectra at 90% power")
plt.ylabel("Instensity (AU)")
plt.xlabel("Wavelength (nm)")
plt.grid(True)
plt.show()
```
